# ðŸŒŸ Exercise 9 :

```python
figure, axis = plt.subplots(2)
x = np.linspace(0,10)
y1 = np.sin(x)
y2 = np.cos(x)
axis[0].set(title="y = sin(x)")
axis[0].plot(x,y1, label="y = sin(x)")
axis[1].set(title="y = cos(x)")
axis[1].scatter(x,y2, label="y = cos(x)")
axis[0].legend()
axis[1].legend()
plt.tight_layout()
```
