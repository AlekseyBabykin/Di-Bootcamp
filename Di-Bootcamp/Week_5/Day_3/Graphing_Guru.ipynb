# Task 1:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0,21,100)
x
```

```python
y = x**2
y
```

```python
figure, axis = plt.subplots()
axis.set(title="Plot of x vs y (y=x^2)", xlabel="X", ylabel="Y")
axis.plot(x,y)
plt.xlim([0,20])
plt.ylim([0,400])
```

# Task 2:

```python
chipo = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep="\t")
chipo
```

```python
chipo["item_price"].apply(lambda x: float(x[1:]))
```

```python
figure, ax = plt.subplots()
ax.set(title="relationship between the price of items and the quantity ordered", xlabel="item_price", ylabel="quantity ordered")
ax.scatter(chipo["item_price"].apply(lambda x: float(x[1:])), chipo["quantity"])
```

```python
figure.savefig("daily.png")
```
