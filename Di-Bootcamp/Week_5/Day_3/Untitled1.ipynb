```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
figure,axes = plt.subplots() 
axes.plot([1,2,3,4],[2,4,5,7])
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
figure,axes = plt.subplots() 
axes.scatter(x,y)
```

```python
data = np.random.rand(1000)
figure,axes = plt.subplots()
axes.hist(data, bins=100)
```

```python
figure,axes = plt.subplots() 
axes.set(title="Line plot", xlabel="X", ylabel="Y")
axes.plot([1,2,3,4],[2,4,5,7], color="pink", linestyle="--")
```

```python
figure,axes = plt.subplots() 
axes.set(title="Line plot", xlabel="X", ylabel="Y")
axes.plot([1,2,3,4],[2,4,5,7], color="pink", linestyle="--", marker="o", mfc="red",mec="blue", label="Alex")

axes.legend()

plt.show()
```

```python
x = np.linspace(-10,10) 
y = x**2
figure,axes = plt.subplots() 
axes.plot(x,y)
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
figure,axes = plt.subplots() 
axes.set(title="y =y**2")
axes.plot(x,y1, color="red", marker="o", linestyle="dashed")
axes.set(title="y =y**1.5")
axes.plot(x,y2, color="blue", linestyle="--", marker="D")
figure.savefig
```
