```python
min_max = {}
for column in data.columns[:-1]:
    min_val = data[column].min()
    max_val = data[column].max()
    min_max[column] = {'min': min_val, 'max': max_val}
    data[column] = (data[column] - min_val) / (max_val - min_val)
data.head()
```

```python
correlations = []
for column in data[:-1]:
 
    correlations.append(data[column].corr(data['medv']))

correlations
sns.heatmap(data.corr())
```

```python
data['age_category'] = pd.cut(data['age'],bins=[0.0,0.333,0.666,0.999],labels=['new','old','very_old'])
data
```

```python
data['future'] = data['rm'] / data['lstat'] 
data
```

# Part II

```python
new_properties = data[data['age'] <= 0.35]
old_properties = data[(data['age'] > 0.35) & (data['age'] <= 0.70)]
very_old_properties = data[data['age'] > 0.70]

figure, axis = plt.subplots(figsize=(10, 6))
axis.hist(new_properties['medv'], bins=10, alpha=0.5, label='New Properties')
axis.hist(old_properties['medv'], bins=10, alpha=0.5, label='Old Properties')
axis.hist(very_old_properties['medv'], bins=10, alpha=0.5, label='Very Old Properties')
plt.xlabel('Median Value (MEDV)')
plt.ylabel('Frequency')
plt.title('Property Age vs. Median Value')
plt.legend()
plt.show()
```

```python
data['CRIM_BIN'] = pd.cut(data['crim'], bins=[0, 0.2, 0.4, 0.6, 0.8, max(data['crim'])], labels=['0-1', '1-5', '5-10', '10-20', '20+'])
```

```python
avg_MEDV_per_CRIM_BIN = data.groupby('CRIM_BIN')['medv'].mean()
```

```python
figure, axis = plt.subplots(figsize=(10, 6))
plt.plot(avg_MEDV_per_CRIM_BIN, marker='o')
plt.xlabel('Binned Crime Rate')
plt.ylabel('avg_MEDV_per_CRIM_BIN')
plt.title('Neighborhood Crime Rate Impact')
plt.show()
```
